```python
import numpy as np
import cv2
import pandas as pd

#'side.mp4'
#'back_so_1.mp4'
#'back_so_2.mp4'
#'front_so_1.mp4'

actions = ['walk', 'run', 'jump', 'legup', 'pickup', 'walk2', 'highfall', 'sitandup', 'movewithchair', 'fallchair', 'upsitbed', 'uplyingbed', 'fallbed']
for a in actions :
    i = 1
    filename = "DATA/Test_data/KinectData/S1P19K" + str(i) + "_" + a + ".mp4"

#filename = 'DATA/Test_data/back_so_2.mp4'
    cap = cv2.VideoCapture(filename)

    if (cap.isOpened() == False):
        print("Error opening the video file")

    current_state = False
    annotation_list = []

    while(True):
        # Read one frame.
        ret, frame = cap.read()
        if not ret:
            break

        # Show one frame.
        cv2.namedWindow('frame', cv2.WINDOW_NORMAL)
        cv2.imshow('frame', frame)

        # Check, if the space bar is pressed to switch the mode.
        key = cv2.waitKey(0)
        if key == 48 :
            current_state = 0
        else :
            current_state = 1

        annotation_list.append(current_state)

    # Convert the list of boolean values to a list of int values.
    print(annotation_list)

    cap.release()
    cv2.destroyAllWindows()

    video = pd.DataFrame()
    video['label'] = annotation_list
    video['frame'] = [(filename.split('/')[-1]).split('.')[0]]*len(video)
    video = video.set_index('frame')

    video.to_csv('DATA/Test_data/KinectData/' + (filename.split('/')[-1]).split('.')[0]+'_label.csv')
```
