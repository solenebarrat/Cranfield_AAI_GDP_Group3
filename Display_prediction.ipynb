# Kinect data prediction display

```python
import numpy as np
import cv2
import pandas as pd

#'side.mp4'
#'back_so_1.mp4'
#'back_so_2.mp4'
#'front_so_1.mp4'

Falls = ["00003","00004","00005","00006","00008","00009","00010","00011","00012","00013","00014","00015","00016"]

for fall_number in Falls :

    #filename = 'KinectData/S1P19K2_fallbed_cut.mp4'

    cap = cv2.VideoCapture('DATA/Falls_MoveNet/Fall_'+fall_number+'_keypoints.mp4')

    if (cap.isOpened() == False):
        print("Error opening the video file")

    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))

    size = (frame_width, frame_height)

    #prediction = pd.read_csv('DATA/Test_data/CRANFIELD_test-data/Harkeys Camera/prediction_MVI_4182_Trim.csv',header=None).to_numpy()[:,0]
    #prediction = pd.read_csv('DATA/Test_data/CRANFIELD_test-data/Harkeys Camera/MVI_4182_Trim_label.csv',index_col = 0).to_numpy()
    #label = pd.read_csv('DATA/Test_data/'+ (filename.split('.')[0]) +'_label.csv').to_numpy()[:,1]
    #prediction = label
    prediction = pd.read_csv('DATA/Falls_MoveNet/Fall_'+fall_number+'_prediction_conv2d.csv',header=None).to_numpy()[:,0]
    accuracy = pd.read_csv('DATA/Falls_MoveNet/Fall_'+fall_number+'_accuracy_conv2d.csv',header=None).to_numpy()[:,0]
    
    #average_prediction = np.zeros(len(prediction))
    #for i in range(len(prediction)) :
        #average_prediction[i] = np.round(np.mean(prediction[i-10 if i-10 >= 0 else i:i+1]))
    #prediction = average_prediction
    
    out = cv2.VideoWriter('DATA/Falls_MoveNet/Fall_'+fall_number+'_prediction_conv2d.mp4', cv2.VideoWriter_fourcc(*'MJPG'),20, size)  #filename.split('.')[0]+

    i = 0
    while(True):
        # Read one frame.
        ret, frame = cap.read()

        if not ret:
            break

        # Show one frame.
        cv2.putText(frame,'Prediction : '+str(np.int64(prediction[i])),(20, 20),cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_4)
        cv2.putText(frame,'Accuracy : '+str((np.round(accuracy[0],1)))+'%',(20, 80),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (100, 200, 0), 2, cv2.LINE_4)

        if i + 1 < len(prediction) :
            i = i+1

        out.write(frame)

        cv2.imshow('frame', frame)


        if cv2.waitKey(1) & 0xFF == ord('s'):
                break


    cap.release()
    out.release()

    cv2.destroyAllWindows()

    print("The video was successfully saved")
```
